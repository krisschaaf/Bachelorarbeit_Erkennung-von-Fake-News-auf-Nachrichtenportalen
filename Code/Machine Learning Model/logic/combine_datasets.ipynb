```python
import pandas as pd
from IPython.display import display
import os

def create_fndg_dataset():
    return pd.read_csv('datasets/fake_news_dataset_german.csv', sep=',', encoding='utf-8')

def create_fang_covid_dataset():
    folder_path = "datasets/FANG-COVID_DATASET/not_preprocessed"
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    all_dfs = []

    for file in csv_files:
        path_to_file = os.path.join(folder_path, file)
        df = pd.read_csv(path_to_file)
        all_dfs.append(df)
        print(f"Loaded: {file} with {len(df)} rows")

    return pd.concat(all_dfs, ignore_index=True)

fndg_dataset = create_fndg_dataset()
display(fndg_dataset)

fang_covid_dataset = create_fang_covid_dataset()
display(fang_covid_dataset)
```

# 📄 Create Testing Dataset for MVP

## 🎯 Goal
Prepare a **testing dataset** for MVP model evaluation.

## 📝 Summary
| Source Dataset       | Fake Samples | Real Samples | Total |
|----------------------|--------------|--------------|-------|
| fang_covid_dataset   | 50           | 50           | 100   |
| fndg_dataset         | 50           | 50           | 100   |
| **Combined Dataset** | **100**      | **100**      | **200** |

## 💾 Output
Saved to:  
`datasets/combined/combined_dataset_200samples_50-50split.csv`

```python
def create_fang_covid_dataset_extract():
    fang_covid_dataset_fake_extract = fang_covid_dataset[fang_covid_dataset['label'] == 'fake'].sample(n=50, random_state=42)
    fang_covid_dataset_true_extract = fang_covid_dataset[fang_covid_dataset['label'] == 'real'].sample(n=50, random_state=42)

    fang_covid_dataset_extract = pd.concat([fang_covid_dataset_fake_extract, fang_covid_dataset_true_extract], ignore_index=True)

    return fang_covid_dataset_extract[['header', 'article', 'label']]

def create_fndg_dataset_extract():
    fndg_dataset_fake_extract = fndg_dataset[fndg_dataset['Fake'] == 1].sample(n=50, random_state=42)
    fndg_dataset_true_extract = fndg_dataset[fndg_dataset['Fake'] == 0].sample(n=50, random_state=42)

    fndg_dataset_extract = pd.concat([fndg_dataset_fake_extract, fndg_dataset_true_extract], ignore_index=True)

    fndg_dataset_extract_reduced = fndg_dataset_extract[['Titel', 'Body', 'Fake']]

    fndg_dataset_extract_reduced_renamed = fndg_dataset_extract_reduced.rename(columns={
        'Titel': 'header',
        'Body': 'article',
        'Fake': 'label'
    })

    fndg_dataset_extract_reduced_renamed['label'] = fndg_dataset_extract_reduced_renamed['label'].map({1: 'fake', 0: 'real'})

    return fndg_dataset_extract_reduced_renamed


fndg_dataset_ex = create_fndg_dataset_extract()
fang_covid_dataset_extract_reduced = create_fang_covid_dataset_extract()

combined_dataset = pd.concat([fang_covid_dataset_extract_reduced, fndg_dataset_ex], ignore_index=True)
combined_dataset_shuffled = combined_dataset.sample(frac=1, random_state=42).reset_index(drop=True)
display(combined_dataset_shuffled)

combined_dataset_shuffled.to_csv('datasets/combined/combined_dataset_200samples_50-50split.csv', index=False, encoding='utf-8')
print("Combined, shuffled dataset saved to 'datasets/combined/combined_dataset_200samples_50-50split.csv'")
```
